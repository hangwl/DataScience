```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv('../datasets/iris_dataset.csv')
```

```python
data = raw_data.copy()
```

```python
data_scaled = preprocessing.scale(data)
```

```python
wcss = []
cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(data_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

elbow_plot = sns.lineplot(x=range(1,cl_num),y=wcss)
elbow_plot.set(xlabel='No. of Clusters', ylabel='WCSS')
plt.show()
```

2 or 3 cluster solutions appear to be the best.

```python
kmeans_2 = KMeans(2)
kmeans_2.fit(data_scaled)
```

```python
clusters_2 = data.copy()
clusters_2['cluster_pred'] = kmeans_2.fit_predict(data_scaled)
```

```python
sns.scatterplot(clusters_2, x='sepal_length', y='sepal_width', hue='cluster_pred')
```

```python
kmeans_3 = KMeans(3)
kmeans_3.fit(data_scaled)
clusters_3 = data.copy()
clusters_3['cluster_pred'] = kmeans_3.fit_predict(data_scaled)
sns.scatterplot(clusters_3, x='sepal_length', y='sepal_width', hue='cluster_pred',palette='muted')
```
